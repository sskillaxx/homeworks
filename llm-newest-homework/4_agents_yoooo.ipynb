```python
import os
from dotenv import find_dotenv, load_dotenv
import asyncio
from agno.agent import Agent
from agno.models.openai.like import OpenAILike
from agno.tools.duckduckgo import DuckDuckGoTools
from agno.workflow import Workflow, Step
from rich.console import Console
from rich.panel import Panel

load_dotenv(find_dotenv())
```

```python
model = OpenAILike(
    id=os.getenv("MODEL_ID"),
    api_key=os.getenv("API_KEY"),
    base_url=os.getenv("BASE_URL")
)

base_instructions = ["Отвечай не больше чем 3 предложениями, раскрой свою личность примерно в 20 словах"]
```

```python
donni = Agent(
    name="Донателло",
    role="умник",
    model=model,
    instructions=[
        "Ты — самый умный и рациональный член команды."
        "Ты обожаешь науку, технологии и изобретения, часто создавая устройства, которые спасают черепашек."
        "Ты предпочитаешь решать проблемы логикой, а не силой.",
        base_instructions[0]
    ]
)
```

```python
rafaelka = Agent(
    name="Рафаэль",
    role="силач",
    model=model,
    instructions=[
         "Ты — самый вспыльчивый и эмоциональный из черепашек." 
         "Ты действуешь импульсивно и часто полагаешься на силу и инстинкты."
         "Ты тяжело переносишь чувство бессилия и несправедливости.",
        base_instructions[0]
    ]
)
```

```python
leo = Agent(
    name="Леонардо",
    role="лидер",
    model=model,
    instructions=[
        "Ты - прирождённый лидер команды и самый дисциплинированный из всех черепашек."
        "Ты чувствуешь огромную ответственность за братьев и часто ставишь долг выше собственных желаний."
        "Иногда ты бываешь слишком серьёзным и строгим, из-за чего тебе бывает сложно расслабиться и проявлять эмоции.",
        base_instructions[0]
    ]
)
```

```python
mikki = Agent(
    name="Микеланджело",
    role="юморист",
    model=model,
    instructions=[
        "Ты — душа команды и главный источник юмора."
        "Ты обожаешь шутки, пиццу и развлечения, стараясь наслаждаться жизнью даже в самых опасных ситуациях."
        "В критические моменты ты способен удивить всех смелостью и неожиданной серьёзностью.",
        base_instructions[0]
    ]
)
```

```python
messages = []

async def get_response(agent, message):
    loop = asyncio.get_running_loop()
    result = await loop.run_in_executor(None, agent.run, message)
    response = result.content
    messages.append(
        {
        agent.name: response
        }
    )
    
    return response
```

```python
async def fighting_shredder():
    first_line = "Черепашки попали в засаду Шредера. Нужно собраться с силами и одолеть врага."
    
    reply_leo = await get_response(leo, first_line)
    print(f"{leo.name}: {reply_leo}\n")

    reply_donni = await get_response(donni, reply_leo)
    print(f"{donni.name}: {reply_donni}\n")

    reply_rafaelka = await get_response(rafaelka, reply_leo)
    print(f"{rafaelka.name}: {reply_rafaelka}\n")

    reply_donni1 = await get_response(donni, reply_rafaelka)
    print(f"{donni.name}: {reply_donni1}\n")

    reply_rafaelka1 = await get_response(rafaelka, reply_donni1)
    print(f"{rafaelka.name}: {reply_rafaelka1}\n")

    reply_mikki = await get_response(mikki, reply_rafaelka1)
    print(f"{mikki.name}: {reply_mikki}\n")

    reply_rafaelka2 = await get_response(rafaelka, reply_mikki)
    print(f"{rafaelka.name}: {reply_rafaelka2}\n")

    final = await get_response(leo, reply_rafaelka2)
    print(f"{leo.name}: {final}")
```

```python
await fighting_shredder()
```
